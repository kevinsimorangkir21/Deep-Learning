1. Buatlah data berupa gambar tungal sebagai input dengan ukuran 124x124 grayscale. Kemudian buatlah model yang terdiri dari dua convolutional layers dan dua pooling layers. Pada convolutional layer pertama, gunakan 32 filter dengan ukuran kernel 5x5, parameter stride (2, 1), dan padding 'valid'. Kemudian, tambahkan pooling layer pertama dengan ukuran 5x5. Selanjutnya, pada convolutional layer kedua, gunakan 64 filter dengan ukuran kernel 5x5, parameter stride (1, 1), dan padding 'same', diikuti oleh pooling layer kedua dengan ukuran 3x3. Setelah melakukan flatten pada output, tambahkan satu fully connected layer dengan 128 neuron yang menggunakan fungsi aktivasi ReLU. Buat output layer menjadi satu neuron dengan fungsi aktivasi softmax. Tampilkan ringkasan model !

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
import matplotlib.pyplot as plt

# Membuat data gambar tunggal berukuran 124x124 grayscale
input_image = np.random.rand(124, 124, 1).astype('float32')

# Membuat model
model = Sequential()

# Convolutional layer pertama
model.add(Conv2D(32, (5, 5), strides=(2, 1), padding='valid',
          activation='relu', input_shape=(124, 124, 1)))
# Pooling layer pertama
model.add(MaxPooling2D(pool_size=(5, 5)))

# Convolutional layer kedua
model.add(Conv2D(64, (5, 5), strides=(1, 1),
          padding='same', activation='relu'))
# Pooling layer kedua
model.add(MaxPooling2D(pool_size=(3, 3)))

# Flatten layer
model.add(Flatten())

# Fully connected layer dengan 128 neuron
model.add(Dense(128, activation='relu'))

# Output layer dengan 1 neuron dan fungsi aktivasi softmax
model.add(Dense(1, activation='softmax'))

# Menampilkan ringkasan model
model.summary()
```

```python
plt.imshow(input_image.squeeze(), cmap='gray')
plt.title('Input Image')
plt.axis('off')
plt.show()
```
